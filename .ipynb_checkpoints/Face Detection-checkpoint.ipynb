```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread("images.jpeg")
```

```python
plt.imshow(img)
plt.show()
```

```python
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = face_detector.detectMultiScale(img, 1.1, 5 )
```

```python
for f in faces:
    x,y,w,h = f
    img_rect = cv2.rectangle(img, (x,y) , (x+w, y+h), (255,255,255), 2)

plt.imshow(img_rect)
plt.show()
```
