```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
```

```python
x
```

```python
x_squared= x**2
```

```python
plt.plot(x, x, marker ="*", color="red")
plt.plot(x, x_squared, marker = "o")
plt.xlabel("numberas 0-9")
plt.ylabel("values")
plt.title("comparision graph")
plt.show()
```
